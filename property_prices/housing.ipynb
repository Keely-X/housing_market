```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Cleaning the initial data. Chosen columns: Suburb, Address, Rooms, Type, Price, Date, Distance, Bathroom, Car, Landsize.

```python
df = pd.read_csv("melb_data.csv")

df = df.drop(columns={'Method', 'SellerG', 'Bedroom2', 'BuildingArea', 'Postcode', 'YearBuilt', 'CouncilArea', 'Lattitude', 'Longtitude', 'Regionname', 'Propertycount'})

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

The Price Distribution: Where do most houses sit?

```python
sns.histplot(df['Price'], bins=50, kde=True,)
plt.title('Distribution of Melbourne Property Prices')
plt.xlim(0, 4000000)
plt.show()
print(f"Mean Price: ${df['Price'].mean():,.2f}")
print(f"Median Price: ${df['Price'].median():,.2f}")
print(f"Mode Price: ${df['Price'].mode()[0]:,.2f}")
```

Which numerical factor is more strongly correlated with price?

```python
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

Geographic Scatter Plot.

```python
plt.figure(figsize=(10, 6))
# Using a scatter plot to show how price drops as distance from CBD increases
sns.scatterplot(x='Distance', y='Price', hue='Type', alpha=0.5, data=df)

plt.title('Melbourne Property Price vs. Distance from CBD')
plt.xlabel('Distance from CBD (km)')
plt.ylabel('Price ($ Millions)')
plt.show()
```
