```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Cleaning the initial data. Chosen columns: Suburb, Address, Rooms, Type, Price, Date, Distance, Postcode, Bathroom, Car, Landsize.

```python
df = pd.read_csv("C:\\Users\\keely\\OneDrive\\Documents\\Coding Projects\\property_prices\\melb_data.csv")

df = df.drop(columns={'Method', 'SellerG', 'Bedroom2', 'BuildingArea', 'YearBuilt', 'CouncilArea', 'Lattitude', 'Longtitude', 'Regionname', 'Propertycount'})

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

The Price Distribution: Where do most houses sit?

```python
sns.histplot(df['Price'], bins=70, kde=True,)
plt.title('Distribution of Melbourne Property Prices')
plt.xlim(0, 4000000)
plt.show()
print(f"Mean Price: ${df['Price'].mean():,.2f}")
print(f"Median Price: ${df['Price'].median():,.2f}")
print(f"Mode Price: ${df['Price'].mode()[0]:,.2f}")
```
